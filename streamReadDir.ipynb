Exemplo de streaming para leitura de arquivos de um diret처rio

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import explode
from pyspark.sql.functions import split

spark = SparkSession \
    .builder \
    .appName("ContadorDePalavras") \
    .getOrCreate()
```

```python
file_dir = spark \
    .readStream \
    .format("text") \
    .option ("inferSchema", "true") \
    .option("maxFilesPerTrigger", 1) \
    .load("./*.txt")
    
```

```python
file_dir.isStreaming
```

```python
words = file_dir.select(
    explode(
        split(file_dir.value, " ")
    ).alias("word")
)
wordCounts = words.groupBy("word").count()
```

```python
from pyspark.sql.functions import desc
wordCounts = wordCounts.sort(desc("count"))
```

```python
query = wordCounts \
    .writeStream \
    .outputMode("complete") \
    .format("console") \
    .start()
    
query.awaitTermination()
```
